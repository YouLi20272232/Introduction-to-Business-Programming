## Part 2: Financial Decision Logic (15 minutes)

### Budget Limit Checker
Apply your if/else skills to real budget management:

````python
print("=== Combined Expense Classifier ===")

# Get inputs (2x input total)
monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter the expense amount: $"))

# --- Absolute amount classification ---
if expense >= 100:
    amount_category = "Major Expense"
elif expense >= 50:
    amount_category = "Moderate Expense"
elif expense >= 10:
    amount_category = "Minor Expense"
else:
    amount_category = "Small Purchase"

# --- Budget percentage classification ---
if monthly_budget == 0:
    budget_message = "Budget is $0 — any expense is a Major purchase."
else:
    share = expense / monthly_budget

    if share > 0.50:
        budget_message = "Major purchase - think carefully"
    elif share > 0.25:
        budget_message = "Significant expense"
    elif share > 0.10:
        budget_message = "Moderate expense"
    else:
        budget_message = "Minor expense - within budget"

# --- Output ---
print("\n=== Results ===")
print("Amount Category:", amount_category)
print("Budget Impact:", budget_message)
````

### Savings Goal Progress
Use multiple conditions to track savings progress:

````python
print("=== Savings Goal Tracker ===")

def safe_float_input(prompt: str) -> float:
    while True:
        raw = input(prompt)
        try:
            value = float(raw)
            return value
        except ValueError:
            print("Please enter a valid number (e.g., 500 or 1250.75).")

goal_amount = safe_float_input("Savings goal: $")
current_savings = safe_float_input("Current savings: $")

# Handle invalid or zero goal to avoid division by zero
if goal_amount <= 0:
    print("\nGoal must be greater than $0 to compute progress.")
    # Still offer a gentle message based on savings
    if current_savings > 0:
        print("Tip: Set a clear goal so your progress can be measured!")
    else:
        print("Set a goal to get started—small steps add up!")
else:
    # Calculate progress percentage (clamp negative savings to 0 for this calc)
    progress = max(current_savings, 0.0) / goal_amount * 100.0

    # Determine feedback message based on your ranges
    if progress >= 100:
        message = "Goal achieved! Consider a new goal."
    elif progress >= 75:
        message = "Almost there! You're doing great."
    elif progress >= 50:
        message = "Halfway there - keep it up!"
    elif progress >= 25:
        message = "Good progress, stay consistent"
    else:
        message = "Just getting started - you've got this!"

    # Output summary
    print("\n=== Progress Summary ===")
    print(f"Goal:             ${goal_amount:,.2f}")
    print(f"Current savings:  ${current_savings:,.2f}")
    print(f"Progress:         {progress:.1f}%")
    print(f"Message:          {message}")
````

## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
print("=== Smart Purchase Advisor ===")

# Get financial information
available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Handle edge cases
if available_money <= 0:
    print("\nYou currently have no available money.")
    print("Advice: Focus on saving before making any purchases.")
else:
    # Calculate affordability ratio (portion of available money)
    affordability_ratio = purchase_price / available_money

    print("\n=== Purchase Analysis ===")

    # 1. Can they afford it?
    if purchase_price > available_money:
        print("⚠️ You cannot afford this purchase right now.")
        print("Advice: Save up first or consider cheaper alternatives.")

    else:
        print("You can afford the item — now let's evaluate if it's a smart decision.")

        # 2. Evaluate how large the purchase is compared to available money
        if affordability_ratio >= 0.75:
            print("This purchase uses MOST of your available money.")
            size_message = "High impact purchase — be cautious."
        elif affordability_ratio >= 0.50:
            print("This purchase uses about HALF of your available money.")
            size_message = "Significant purchase — think carefully."
        elif affordability_ratio >= 0.25:
            print("This purchase uses about a QUARTER of your available money.")
            size_message = "Moderately sized — plan your budget."
        else:
            print("This is a small portion of your available money.")
            size_message = "Low impact — financially manageable."

        # 3. Consider if the purchase makes sense compared to income
        if purchase_price > monthly_income:
            income_message = "The item costs more than your monthly income — this is a BIG purchase."
        elif purchase_price > monthly_income * 0.5:
            income_message = "The item is a large fraction of your monthly income."
        elif purchase_price > monthly_income * 0.25:
            income_message = "The item is moderately sized compared to your income."
        else:
            income_message = "The item is well within your monthly income level."

        # 4. Provide final advice
        print("\n=== Smart Advice ===")
        print(size_message)
        print(income_message)

        if affordability_ratio < 0.25 and purchase_price < monthly_income * 0.25:
            print("Overall: 👍 This seems like a reasonable purchase.")
        elif affordability_ratio >= 0.75 or purchase_price > monthly_income:
            print("Overall: ⚠️ This may not be the best time to buy — reconsider.")
        else:
            print("Overall: 🤔 It's possible, but think it over before buying.")
````

### Sample Decision Logic to Implement

- **Can't afford**: "Save up first - you need $X more"
- **Can afford but expensive**: "You can afford it, but it's Y% of your available money. Consider waiting."
- **Reasonable purchase**: "This seems like a reasonable purchase for your budget"
- **Small purchase**: "This is a minor expense - go for it!"

## Part 4: Building Your Finance Tracker Foundation (15 minutes)

### Start Your Semester Project

Create a new notebook called "Personal_Finance_Tracker" and build the foundation:

````python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# This week's feature: Expense categorisation
print("\n--- Expense Entry ---")
expense_amount = float(input("Enter an expense: $"))
expense_description = input("What was this expense for? ")

# Categorisation rules (you can modify next week!)
if expense_amount >= 100:
    category = "Major Expense"
    feedback = "This is a large expense. Make sure it fits within your budget."
elif expense_amount >= 50:
    category = "Moderate Expense"
    feedback = "A moderate cost—manageable, but track these carefully."
elif expense_amount >= 10:
    category = "Minor Expense"
    feedback = "A small expense. Usually fine, but they can add up over time."
else:
    category = "Tiny Purchase"
    feedback = "Very small cost—little impact on your budget."

# Compare expense to income (extra helpful insight)
income_ratio = expense_amount / monthly_income

if income_ratio > 0.5:
    extra = "⚠️ This is more than half of your monthly income—reconsider!"
elif income_ratio > 0.25:
    extra = "This is a noticeable part of your income."
elif income_ratio > 0.10:
    extra = "This takes a small but meaningful portion of your income."
else:
    extra = "This has very little impact on your income."

# Output summary
print("\n=== Expense Summary ===")
print(f"Description: {expense_description}")
print(f"Amount: ${expense_amount:.2f}")
print(f"Category: {category}")
print(f"Feedback: {feedback}")
print(f"Income Impact: {extra}")

print("\nThanks for logging your expense! More features coming next week.")
````
